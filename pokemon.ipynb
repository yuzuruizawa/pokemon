```python
#media HP 
meanAtt =np.mean(df.Attack)
sdAtt = np.std(df.Attack)
meanHP = np.mean(df.HP)
sdHP = np.std(df.HP)
```

```python
import matplotlib.patches as patches
import matplotlib.pyplot as plt

fig, graph = plt.subplots()
graph.scatter(df.Attack,df.HP)

graph.scatter(meanAtt,meanHP)

margemErro = patches.Ellipse([meanAtt,meanHP],sdAtt*2,sdHP*2, alpha=0.25)
graph.add_patch(margemErro)

plt.xlabel('Ataque')
plt.ylabel('HP')
```

```python
from sklearn.cluster import KMeans
attack_hp = np.column_stack((df.Attack,df.HP))

km = KMeans(n_clusters= 7).fit(attack_hp)

km.cluster_centers_
```

```python
clusters = km.cluster_centers_

plt.scatter(df.Attack,df.HP)
plt.scatter(clusters[:,0],clusters[:,1])
```
